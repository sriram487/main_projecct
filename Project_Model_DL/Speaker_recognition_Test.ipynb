```python
import json
import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow.keras as keras
import matplotlib.pyplot as plt
```

```python
data_test_path = r"C:\Users\sriram\Desktop\Main_Project\Project_Data\data_1Test.json"
model_path = r"C:\Users\sriram\Desktop\Main_Project\Project_Model_DL\model_CNN_4c_V2.h5"
```

```python
model = keras.models.load_model(model_path)
```

```python
def load_data(data_path):
    """Loads training dataset from json file.
        :param data_path (str): Path to json file containing data
        :return X (ndarray): Inputs
        :return y (ndarray): Targets
    """

    with open(data_path, "r") as fp:
        data = json.load(fp)

    X = np.array(data["mfcc"])
    y = np.array(data["labels"])
    return X, y
```

```python
X_test , y_test = load_data(data_test_path)
```

```python
y_pred = []
for i in X_test:
    i = i[np.newaxis, ...]
    prediction = model.predict(i)
    pred = np.argmax(prediction, axis=1)
    y_pred.append(pred[0])
    
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
plt.imshow(cm, cmap='binary')
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix


print(f1_score(y_test, y_pred, average="macro"))
print(precision_score(y_test, y_pred, average="macro"))
print(recall_score(y_test, y_pred, average="macro"))
```
